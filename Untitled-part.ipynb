```python
def plan():
    displayP = f'\n--- Plan Material\n'
    menuPlan = ['Planning', 'Back']
    prompt2 = pps.inputMenu(
        prompt=displayP, choices=menuPlan, numbered=True)
    if prompt2 == 'Planning':
        inputCode = int(input('Enter new material code: '))
        addValue = [value for databaru in listMaterial for value in databaru]
        if inputCode in addValue:
            print('\nMaterial already exist')
        else:
            newCode = input('Please enter the code: ')
            newMaterial = input('Please enter the material: ')
            newInventory = int(input('Please enter the inventory amount: '))
            newDistribution = input('Please enter the distribution plant: ')
            newSupplier = input('Please enter the supplier: ')
            newListMaterial = [{
                'Code' : newCode,
                'Material' : newMaterial,
                'Inventory' : newInventory,
                'Distribution' : newDistribution,
                'Supplier' : newSupplier
            }]
            template(listMaterial)
            save = input('Save the data?(Y/N)').lower()
            if save == 'y':
                listMaterial.extend(newListMaterial)
                template(listmaterial)
                print('New data has saved')
            else:
                print('New data has not saved')
    elif prompt2 == 'Back':
        main()
    plan()
```
